### Multicollinearity check

For gradient boosted trees, there's generally no strong need to check for multicollinearity because of its robustness. But practically speaking, you still should do some basic checks. For example, if you discover that two variables are 100% the same, then of course there's no point in keeping both. Even if it's 98% correlated, it's usually okay to drop one variable without degrading the overall model. Collinearity may mess up feature importance.

```python
# multicollinearity
# matrix of a pearson product-moment correlation coefficients
X_cs_np_corr = np.corrcoef(cs_dat, rowvar=0)
```

```python
# create a mask to fill upper triangle and diagonal with 0, lower triangle are 1
mask_cs = np.tri(X_cs_np_corr.shape[0], k = -1)
```

```python
# mask out the lower triangle
X_cs_np_corr_masked = np.ma.array(X_cs_np_corr, mask=mask_cs) 
```

```python
fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(111)
plot = ax1.matshow(X_cs_np_corr_masked, aspect = 1)
fig.colorbar(plot, orientation='vertical', shrink = 0.8)
labels = cs_dat.columns.values
ax1.set(xticks=np.arange(len(labels)), xticklabels=labels,
       yticks=np.arange(len(labels)), yticklabels=labels)
plt.title('Correlation matrix, CS')
plt.savefig('fig/corrmatrix-CS-present.eps', format='eps', bbox_inches='tight')
```

```python
# remove variables that cause multicollinearity

# CS: remove lcop1, lcop2, cumf
X_cs = cs_dat[['cfin1', 'cfin2', 'chel1', 'chel2','fbar','recr', 'totaln', 'sst', 'sal', 'xmonth', 'nao']].copy()
y_cs = cs_dat[['length']]

# extract col names
cs_names = list(X_cs.columns.values) 

# Convert both DataFrame objects to a NumPy array (feature matrices) 
X_cs_np = X_cs.astype(np.float).values
y_cs_np = y_cs.astype(np.float).values

#return a flattened array y
y_cs_np = y_cs_np.ravel() 
```

### 1. Data partitioning: 3-way split


```python
"""
Divide data into train (0.5), test (0.25), and validation (0.25) data 
train - to fit the models
val -  to estimate prediction error for model selection
test - to assess the generalization error of the final chosen model
"""

X_cs_train, X_cs_test, y_cs_train, y_cs_test = cv.train_test_split(X_cs_np, y_cs_np, test_size=0.5)
X_cs_test, X_cs_val, y_cs_test, y_cs_val = cv.train_test_split(X_cs_test, y_cs_test, test_size=0.5)
```

```python
X_cs_train.shape, X_cs_val.shape,  X_cs_test.shape
#X_nw_train.shape, X_nw_val.shape,  X_nw_test.shape
```

### Model fitting

```python
%%time

params = {'n_estimators': 500,
          'max_depth': 6,
          'min_samples_leaf': 4,
          'learning_rate': 0.05,
          'min_samples_split': 1, 
          'subsample': 0.75,
          'random_state': 42, 
          'loss': 'ls'} 

est_cs = ensemble.GradientBoostingRegressor(**params)
est_cs.fit(X_cs_train, y_cs_train)
```

```python
from pickle import dumps
print("CS model size: %0.1f MB" % (len(dumps(est_cs)) / 1e6))
```

### Model estimation

```python
# Print the score (R^2) 
acc_cs_train = est_cs.score(X_cs_train, y_cs_train)
acc_cs_val = est_cs.score(X_cs_val, y_cs_val)
acc_cs_test = est_cs.score(X_cs_test, y_cs_test)
```

```python
print('Coefficients of determination CS')
print('R^2 train: %.4f' % acc_cs_train)
print('R^2 val: %.4f' % acc_cs_val)
print('R^2 test: %.4f' % acc_cs_test)
```

```python
# mse observed - predicted
mse_cs = mean_squared_error(y_cs_test, est_cs.predict(X_cs_test))

print('MSE CS: %.4f' % mse_cs)
```

```python
%%time
# compare to baseline model - simple regression
lm_cs = LinearRegression().fit(X_cs_train, y_cs_train)

print("CS MSE train: %0.3f, test: %0.3f" %
      (mean_squared_error(y_cs_train, lm_cs.predict(X_cs_train)),
       mean_squared_error(y_cs_test, lm_cs.predict(X_cs_test))))
```

```python
plt.figure(figsize=(12,12))

fig, ax = plt.subplots()

plt.scatter(y_cs_test, est_cs.predict(X_cs_test), color='b', marker='^', alpha=.1)
plt.title('True vs predicted fish length')
plt.grid(True)
ax.plot([y_cs_test.min(), y_cs_test.max()], [y_cs_test.min(), y_cs_test.max()], 'k--', lw=3)
ax.set_xlabel('True length, cm')
ax.set_ylabel('Predicted length, cm')

plt.savefig('fig/true-pred-present.eps', format='eps')
```

```python
# compute test set deviance
test_cs_score = np.zeros((params['n_estimators'],), dtype=np.float64)
```

```python

for i, y_cs_pred in enumerate(est_cs.staged_predict(X_cs_test)): 
   test_cs_score[i] = est_cs.loss_(y_cs_test, y_cs_pred)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.title('Deviance CS')
plt.plot(np.arange(est_cs.n_estimators) + 1,
        est_cs.train_score_, 'b-',
        label = 'Training')
plt.plot(np.arange(est_cs.n_estimators) + 1,
       test_cs_score, 'r-',
        label = 'Test')
plt.legend(loc = 'right') 
plt.xlabel('Boosting Iterations')
plt.ylabel('MSE')
plt.savefig('fig/devianceCS-present.eps', format='eps', bbox_inches='tight')
```

### Feature importance

```python
fx_imp_cs = pd.Series(est_cs.feature_importances_, index=list(X_cs.columns.values))
fx_imp_cs /= fx_imp_cs.max()
fx_imp_cs.sort_values(inplace=True)
fx_imp_cs.plot(kind='barh', figsize = (8,5))
plt.savefig('fig/featureCS-present.eps', format='eps', bbox_inches='tight')
```

```python
# partial dependence plot
from sklearn.ensemble.partial_dependence import plot_partial_dependence
import itertools

#take top 3 features and produce combinations
fx_imp_cs3 = list(fx_imp_cs.sort_values(ascending=False)[:3].index)
features = []
for i in xrange(1, len(fx_imp_cs3)):
    els = [list(x) for x in itertools.combinations(fx_imp_cs3, i)]
    features.extend(els)
    
fig, axs = plot_partial_dependence(est_cs, 
                                   X_cs_train, 
                                   features, 
                                   feature_names=list(X_cs.columns.values),
                                   n_cols=3,
                                   figsize=(10,5)
                                   )
fig.suptitle('Partial dependence of herring length on features\n'
                 'for the CS')
plt.subplots_adjust(top=0.9)
plt.savefig('fig/partdepCS-present.eps', format='eps', bbox_inches='tight')
```

```python
features = list(fx_imp_cs.sort_values(ascending=False)[3:].index)
  
fig, axs = plot_partial_dependence(est_cs, 
                                   X_cs_train, 
                                   features, 
                                   feature_names=list(X_cs.columns.values),
                                   n_cols=3,
                                   figsize=(15,10)
                                   )
fig.suptitle('Partial dependence of herring length on features\n'
                 'for the CS')
plt.subplots_adjust(top=0.9)
plt.savefig('fig/partdepCS-tuned-all-present.eps', format='eps', bbox_inches='tight')
```

### Early stopping

```python
# use early stopping to find the number of trees

patience = 5
gbr_cs = ensemble.GradientBoostingRegressor(
    n_estimators=1, 
    max_depth=6, 
    learning_rate=0.05,
    min_samples_split=1, 
    subsample=0.75,
    random_state=42, 
    loss='ls',
    warm_start=True)


best_error = np.inf

for i in range(2000):
    gbr_cs.set_params(n_estimators=i + 1)
    gbr_cs.fit(X_cs_train, y_cs_train)
    val_error = mean_squared_error(y_cs_test, gbr_cs.predict(X_cs_test))
    if val_error < best_error:
        best_error = val_error
    else:
        patience -= 1
        if patience < 0:
            print('Converged at tree #%d' % (i + 1))
            break

```

### Tuning with grid search

Tuning model with grid search. Exhaustive search over specified parameter values for an estimator.

Grid search takes a set of possible values for each hyperparameter that should be tuned and evaluates a model trained on each element of the Cartesian product of the sets. It is an exhaustive search that trains and evaluates a model for each possible combination of the hyperparameter values supplied.


```python
# tune the most important once
param_tune = {'max_depth': [4, 6, 8],
              'learning_rate': [0.1, 0.05, 0.01],
              }

est_cs_tune = ensemble.GradientBoostingRegressor(n_estimators=500)

gs_cv_cs_tune = GridSearchCV(est_cs_tune,
                        param_tune, 
                        scoring='mean_squared_error', 
                        n_jobs=4
                       )

gs_cv_cs_tune.fit(X_cs_train, y_cs_train)

print('Best hyperparameters: %r' % gs_cv_cs_tune.best_params_)
```

```python
# add extra parameters 
gs_cv_cs_tune.best_params_['random_state'] = 42
gs_cv_cs_tune.best_params_['subsample'] = 0.75
```

### Model fitting

```python
est_cs_tune.set_params(**gs_cv_cs_tune.best_params_)
```

```python
est_cs_tune.fit(X_cs_train, y_cs_train)
```
